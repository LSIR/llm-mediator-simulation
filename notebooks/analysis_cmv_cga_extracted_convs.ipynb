```python
import os
import pandas as pd

datapath = "../data/reddit/cmv"

filenames = list(os.listdir(datapath))
filenames.sort()

thread_len_distribution = []

for filename in filenames:
    if filename.endswith(".csv"):
        submission_id = filename.split("-")[0].split("_")[1]

        # open the csv file
        df = pd.read_csv(os.path.join(datapath, filename))

        thread_len_distribution.append(len(df) - 2)  # trucated with last 2 comments

# plot the distribution
import matplotlib.pyplot as plt

plt.hist(thread_len_distribution, bins=100)
plt.title(
    f"Distribution of truncated thread lengths ({len(thread_len_distribution)} threads)"
)
plt.xlabel("Thread length")
plt.ylabel("Number of threads")
plt.xlim(3, 15)
plt.ylim(0, 50)
plt.grid()
plt.show()
```
