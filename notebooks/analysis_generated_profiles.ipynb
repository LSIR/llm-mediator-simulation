```python
import json
import matplotlib.pyplot as plt


with open("../data/reddit/cmv/reddit_user_profiles.json", "r") as f:
    debater_profiles = json.load(f)

submission_num_distribution = []
confidence_distribution = []
for profile in debater_profiles.values():
    submission_num_distribution.append(profile["submission_num"])
    confidence_distribution.append(profile["confidence"])

# plot the distributions with matplotlib


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(submission_num_distribution, bins=50, color="blue", alpha=0.7)
plt.title(f"Submission Number Distribution ({len(submission_num_distribution)} users)")
plt.xlabel("Number of Submissions")
plt.ylabel("Frequency")
plt.subplot(1, 2, 2)
plt.hist(confidence_distribution, bins=50, color="green", alpha=0.7)
plt.title(f"Confidence Distribution ({len(submission_num_distribution)} users)")
plt.xlabel("Confidence")
plt.ylabel("Frequency")
plt.tight_layout()
```
